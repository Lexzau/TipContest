<h1>Daten holen um Sie Lokal zu speichern

```python
import urllib.request


#Get all Teams
for y in year:
    with urllib.request.urlopen(get_all_teams + y) as url:
        s = url.read()
        with open("Allteams_from_" + y + '.json', 'w') as d:
            d.write(str(s))
            
#Get all Matches
for y in year:
    with urllib.request.urlopen(get_all_matches + y) as url:
        s = url.read()
        with open("Allgames_from_" + y + '.json', 'w') as d:
            d.write(str(s))
```
